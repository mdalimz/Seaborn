```python
import seaborn as ali
```

```python
penguins = ali.load_dataset("penguins")
penguins
```

# Distribution plots

# Figure-level vs. axes-level functions


# displot

```python
ali.displot(data=penguins, x="flipper_length_mm", kind="kde")
ali.displot(data=penguins, x="flipper_length_mm", kind="ecdf")
ali.displot(data=penguins, x="flipper_length_mm", kde=True)
ali.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm")
ali.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm", kind="kde")
ali.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm", kind="kde", rug=True)
ali.displot(data=penguins, x="flipper_length_mm", hue="species", kind="kde")
ali.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
ali.displot(data=penguins, x="flipper_length_mm", hue="species", col="sex", kind="kde")
ali.displot(data=penguins, y="flipper_length_mm", hue="sex", col="species",kind="ecdf", height=4, aspect=.7,)
data=ali.displot(data=penguins, y="flipper_length_mm", hue="sex", col="species",kind="kde", height=4, aspect=.7,)
data.set_axis_labels("Density (a.u.)", "Flipper length (mm)")
data.set_titles("{col_name} penguins")
```
